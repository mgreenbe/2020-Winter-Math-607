### Density Estimation

Estimate the densities for:

1. the heights of women,
2. the heights of men, and
3. the heighs of both women and men.

Do this
1. nonparametrically, using histograms, and
2. parametrically, by assuming the distribution of the heights of women and the distribution of the heights of men are both Gaussian.

```python
m_male = np.mean(male_heights)
s_male = np.std(male_heights)
m_female = np.mean(female_heights)
s_female = np.std(female_heights)
m_all = np.mean(heights)
s_all = np.std(heights)
n_male = len(male_heights)
n_female = len(female_heights)
p_male = n_male/(n_male + n_female)
p_female = n_female/(n_male + n_female)
```

```python
def g(x, mu, sigma):
    return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-0.5*(x - mu)**2/sigma**2)
```

```python
_, bins = np.histogram(heights, bins=20, density=True);

plt.figure(figsize=(10, 5))

plt.hist(female_heights,
         bins=bins, density=True, color=(r0, g0, b0, 0.9), edgecolor='black')
plt.hist(male_heights,
         bins=bins, density=True, color=(r1, g1, b1, 0.7), edgecolor='black');
plt.hist(heights,
         bins=bins, density=True, color=(r2, g2, b2, 0.5), edgecolor='black');

xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)

y_male = g(x, m_male, s_male)
y_female = g(x, m_female, s_female)
y_all = p_male*y_male + p_female*y_female # mixture

plt.plot(x, y_male, 'k-', lw=5)
plt.plot(x, y_male, 'k-', lw=2, color=(r1, g1, b1))
plt.plot(x, y_female, 'k-', lw=5)
plt.plot(x, y_female, 'k-', lw=2, color=(r0, g0, b0))
plt.plot(x, y_all, 'k-', lw=5)
plt.plot(x, y_all, 'k-', lw=2, color=(r2, g2, b2))
plt.plot(x, y, 'r--', lw=4)
plt.plot(x, g(x, m0, s0), 'r--', lw=4)
plt.plot(x, g(x, m1, s1), 'r--', lw=4)
```

## Supervised Learning

```python
df.Gender = df.Gender.apply(lambda x: 0 if x == "Female" else 1)
df = df.sample(frac=1).reset_index(drop=True)
df
```

### Train/Test Split

```python
X_train = df[:9000].Height.values.reshape(-1, 1)
y_train = df[:9000].Gender.values
X_test = df[9000:].Height.values.reshape(-1, 1)
y_test = df[9000:].Gender.values
```

### The Bayes Classifier

$$
\begin{aligned}
x&\in\mathbb{R}&&\text{height}\\
\\
z&\in\{\text{male, female}\}&&\text{class label}\\
\\
p(z=i|x) &= \frac{p(x|z=i)p(z=i)}{p(x)}&&\text{Bayes' Theorem}\\
&= \frac{p(x|z=i)p(z=i)}{p(x|z=0)p(z=0) + p(x|z=1)p(z=1)}&&\text{by the Law of Total Probability}
\end{aligned}
$$

```python
FEMALE = 0
MALE = 1

X_female = X_train[y_train == FEMALE]
X_male = X_train[y_train == MALE]

bins = np.arange(np.floor(np.min(X_train)), np.ceil(np.max(X_train)))

hist_female, bins_female = np.histogram(X_female, bins=bins, density=True)
hist_male, bins_male = np.histogram(X_male, bins=bins, density=True)
```

```python
bins = bins.reshape(1, -1)

p_female = len(X_female)/(len(X_female) + len(X_male))
p_male = len(X_male)/(len(X_female) + len(X_male))

p_x_female = hist_female[np.argmin(X_test > bins, axis=1) - 1]
p_x_male = hist_male[np.argmin(X_test > bins, axis=1) - 1]

p_x = p_x_female*p_female + p_x_male*p_male
p_female_x = p_x_female*p_female/p_x
p_male_x = p_x_male*p_male/p_x
```

```python
y_pred = 1.0*(p_male_x > p_female_x)
```

```python
np.sum(y_pred == y_test)/len(y_test)
```
